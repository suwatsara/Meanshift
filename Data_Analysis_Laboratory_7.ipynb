```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from datetime import timedelta
# if using a Jupyter notebook, includue:
%matplotlib inline

import sys
```

```python



data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/customers.csv")
print(data.head())
```

```python
data["Income"] = data[["Annual Income (k$)"]]
data["Spending"] = data[["Spending Score (1-100)"]]
data = data[["Income", "Spending"]]
print(data.head())
```

```python
from sklearn.cluster import MeanShift
model = MeanShift(bandwidth=10)
model.fit(data)
pred = model.fit_predict(data)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
plt.scatter(data["Income"], data["Spending"], c=pred, cmap='rainbow', alpha=0.9)
plt.show()
```

```python
df = pd.read_csv('fishdata.csv',encoding= 'unicode_escape')
df.head()
```

```python
df.rename(columns={'2020':'Year', '6':'Month', '18':'Day', '14':'Hour', '30':'Minute', '33':'Second',
       'qCBrs9 R! cFInitializing.... Please Wait!':'Carbonmonoxide', 'Unnamed: 7':'PM2.5',
       'Unnamed: 8':'External_Temp', 'Unnamed: 9':'Humidity', 'Unnamed: 10':'Internal_Temp', 'Unnamed: 11':'Ignore'
},inplace = True)
```

```python
pd.reset_option('display.float_format')
pd.options.display.float_format = '{:.2f}'.format
```

```python
df.dropna(inplace=True)
```

```python
df2 = df[0:40000]
```

```python

from sklearn.cluster import MeanShift
model = MeanShift(bandwidth=5)
model.fit(df2)
pred = model.fit_predict(df2)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
plt.scatter(df2["External_Temp"], df2["Humidity"], c=pred, cmap='plasma', alpha=0.9)
plt.show()
```

```python
from sklearn.cluster import MeanShift
model = MeanShift(bandwidth=10)
model.fit(df2)
pred = model.fit_predict(df2)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
plt.scatter(df2["External_Temp"], df2["Humidity"], c=pred, cmap='plasma', alpha=0.9)
plt.show()
```
